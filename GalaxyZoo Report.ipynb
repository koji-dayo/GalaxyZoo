```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
ellip = pd.read_csv('DataCSV/Elliptical.csv')
spiral = pd.read_csv('DataCSV/Spiral.csv')
```

```python
k_list = list(range(0, 100, 5))
xlabel = 'K'
ylabel = 'Quantity'
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.hist(ellip['K'], k_list,histtype="step",label='Elliptical')
plt.hist(spiral['K'], k_list,histtype="step",label='Spiral')
plt.legend()
```

```python
#GLON Plot
glon_list = list(range(0, 100, 5))
xlabel = 'GLON'
ylabel = 'Quantity'
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.hist(ellip['glon'], glon_list,histtype="step",label='Elliptical')
plt.hist(spiral['glon'], glon_list,histtype="step",label='Spiral')
plt.legend()
```
